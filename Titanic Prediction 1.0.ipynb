# Data Preprocessing

```python
titanic = titanic.drop(['Cabin'], axis = 1)
titanic_test = titanic_test.drop(['Cabin'], axis = 1)
titanic.head()
```

```python
titanic = titanic.drop(['PassengerId','Name','Ticket'], axis=1)
titanic_test    = titanic_test.drop(['Name','Ticket'], axis=1)
```

Dropping Embarked as it has not relevance to who survived

```python
titanic.drop(['Embarked'], axis=1,inplace=True)
titanic_test.drop(['Embarked'], axis=1,inplace=True)
```

Only for Test data, Filling the fare with Median values 

```python
titanic_test["Fare"].fillna(titanic_test["Fare"].median(), inplace=True)
```

Converting fare from float to int 

```python
titanic['Fare'] = titanic['Fare'].astype(int)
titanic_test['Fare'] = titanic_test['Fare'].astype(int)
```

Getting fares of people who survived and didn't survive! 

```python
fare_survived = titanic['Fare'][titanic['Survived']==1]
fare_not_survived = titanic['Fare'][titanic['Survived']==0]
```

```python
average_age_titanic   = titanic["Age"].mean()
std_age_titanic       = titanic["Age"].std()
count_nan_age_titanic = titanic["Age"].isnull().sum()
print(average_age_titanic)   
print(std_age_titanic)
print(count_nan_age_titanic)
```

```python
average_age_titanic_test   = titanic_test["Age"].mean()
std_age_titanic_test       = titanic_test["Age"].std()
count_nan_age_titanic_test = titanic_test["Age"].isnull().sum()
print(average_age_titanic_test)   
print(std_age_titanic_test)
print(count_nan_age_titanic_test)
```

```python
# Generate Random numbers between Mean - Std and Mean + Std
rand_titanic = np.random.randint(average_age_titanic - std_age_titanic,average_age_titanic + std_age_titanic, count_nan_age_titanic)
rand_titanic_test = np.random.randint(average_age_titanic_test - std_age_titanic_test,average_age_titanic_test + std_age_titanic_test, count_nan_age_titanic_test)
len(rand_titanic)
len(rand_titanic_test)
```

```python
#Drop all null values and convert to int 
titanic['Age'].dropna().astype(int)
titanic_test['Age'].dropna().astype(int)
```

```python
#fill all NaN Values with random numbers generated before 
titanic['Age'][np.isnan(titanic['Age'])] = rand_titanic
titanic_test['Age'][np.isnan(titanic_test['Age'])] = rand_titanic_test
```

```python
#Convert it to int 
titanic['Age'] = titanic['Age'].astype(int)
titanic_test['Age'] = titanic_test['Age'].astype(int)
```

```python
# Creating a new variable called Family (Consists of Parents and Siblings)
titanic['Family'] = titanic['Parch'] + titanic['SibSp']
titanic['Family'].loc[titanic['Family']>0] = 1
titanic['Family'].loc[titanic['Family']==0] = 0
```

```python
#Creating the new variable in test 
titanic_test['Family'] = titanic_test['Parch'] + titanic_test['SibSp']
titanic_test['Family'].loc[titanic_test['Family']>0] = 1
titanic_test['Family'].loc[titanic_test['Family']==0] = 0
```

```python
#dropping both Parch and Sibsp
titanic = titanic.drop(['SibSp', 'Parch'], axis = 1)
titanic_test = titanic_test.drop(['SibSp','Parch'], axis = 1)
```

Gender - Important as Children and Females have higher chances of survival 


```python
# We Can see that childrn (Age<12) have higher chances of survival 
def get_person(passenger):
    age,sex = passenger
    return 'child' if age < 12 else sex 
```

```python
titanic['Person'] = titanic[['Age','Sex']].apply(get_person, axis =1 )
titanic_test['Person'] = titanic[['Age','Sex']].apply(get_person, axis =1 )
```

```python
titanic
```

```python
titanic_test
```

```python
titanic.drop(['Sex'], axis =1 , inplace = True)
titanic_test.drop(['Sex'], axis =1 , inplace = True)
```

Converting data set with strings into columns and numbers!


```python
person_dummies_titanic  = pd.get_dummies(titanic['Person'])
person_dummies_titanic.columns = ['Child','Female','Male']
person_dummies_titanic.drop(['Male'], axis=1, inplace=True)

person_dummies_test  = pd.get_dummies(titanic_test['Person'])
person_dummies_test.columns = ['Child','Female','Male']
person_dummies_test.drop(['Male'], axis=1, inplace=True)
```

```python
titanic = titanic.join(person_dummies_titanic)
```

```python
titanic_test = titanic_test.join(person_dummies_test)
```

```python
titanic.head()
```

```python
titanic.drop(['Person'], axis = 1, inplace = True)
titanic_test.drop(['Person'], axis = 1, inplace = True)
```

```python
pclass_dummies_titanic  = pd.get_dummies(titanic['Pclass'])
pclass_dummies_titanic.columns = ['Class_1','Class_2','Class_3']
pclass_dummies_titanic.drop(['Class_3'], axis=1, inplace=True)

pclass_dummies_test  = pd.get_dummies(titanic_test['Pclass'])
pclass_dummies_test.columns = ['Class_1','Class_2','Class_3']
pclass_dummies_test.drop(['Class_3'], axis=1, inplace=True)
```

```python
titanic.drop(['Pclass'], axis =1 , inplace = True)
titanic_test.drop(['Pclass'], axis =1, inplace = True)
```

```python
titanic = titanic.join(pclass_dummies_titanic)
titanic_test = titanic_test.join(pclass_dummies_test)
```

Defining the training and test data set

```python
X_train = titanic.drop(['Survived'], axis= 1)
y_train = titanic['Survived']
X_test = titanic_test.drop(['PassengerId'], axis = 1).copy() 
```

Logistic regression 

```python
from sklearn.linear_model import LogisticRegression
```

```python
logr = LogisticRegression()
logr.fit(X_train, y_train)
y_pred = logr.predict(X_test)
```

```python
logr.score(X_train, y_train)
```

Random forrest classifier

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
randomf = RandomForestClassifier()
randomf.fit(X_train, y_train)
y_pred = randomf.predict(X_test)
```
